Q. 2) In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph.

```python
data["Date_N"]=pd.to_datetime(data["Release_Date"].str.strip())
```

```python
data.head()
```

```python
data_year=data["Date_N"].dt.year.value_counts().sort_index()
data_year
```

```python
data_year.plot(kind="bar")
```

```python
# Convert to DataFrame
data_year = data_year.reset_index()
data_year.columns = ['release_year', 'count']
data_year
```

```python
# Create line chart
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x='release_year', y='count', data=data_year)

plt.title('Content Released per Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()
```
